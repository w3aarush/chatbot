```python
import pandas as pd
import chromadb as chroma
```

```python
chat_df = pd.read_csv('chatdata.csv')
chat_df.head()
```

```python
chat_df.info()
```

```python
documents = chat_df['user_input'].tolist()
ids = chat_df['id'].astype(str).tolist()
metadata = chat_df.drop(columns=['user_input']).to_dict(orient='records')
```

```python
metadata
```

```python
documents
```

```python
ids
```

## Database Creation

```python
client = chroma.PersistentClient(path="./chatdatabase")
```

```python
client.database
```

```python
collection = client.get_or_create_collection(name="chat_collection")
```

```python
collection.add(
    documents=documents,
    ids=ids,
    metadatas=metadata
)
```

```python
collection.count()
```

```python
dir(collection)
```

```python
collection.name
```

### Testing

```python
result = collection.query(
    query_texts=["What is the weather like today?"],
    n_results=2
)
```
